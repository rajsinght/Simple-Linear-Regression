```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('Years Experience and Salary.csv')
```

```python
real_x= data.iloc[:,0].values
real_y= data.iloc[:,1].values
real_x= real_x.reshape(-1,1)
real_y= real_y.reshape(-1,1)
```

# Train the Model

```python
train_x,test_x,train_y,test_y=train_test_split(real_x,real_y,test_size=0.3,random_state=0)
```

# Applying Linear Regression

```python
Lin= LinearRegression()
Lin.fit(train_x,train_y)
```

```python
pred_y=Lin.predict(test_x)
```

# Testing the model by using 3rd index of the data set

```python
pred_y[3]
```

```python
test_y[3]
```

# Plotting the data

```python
plt.scatter(train_x,train_y,color="green")
plt.plot(train_x,Lin.predict(train_x),color='blue')
plt.title('Salary & Experience graph')
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(test_x,test_y,color="green")
plt.plot(train_x,Lin.predict(train_x),color='blue')
plt.title('Salary & Experience graph')
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```
